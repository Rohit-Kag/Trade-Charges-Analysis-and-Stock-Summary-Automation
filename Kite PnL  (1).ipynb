```python
import numpy as np
import pandas as pd 
```

```python
df = pd.read_csv('orders.csv')
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Qty.']
```

```python
df['Qty.'] = df['Qty.'].str.split('/').str[0].astype(int)
```

```python
df
```

```python
df.drop(df[df['Product'] != 'MIS'].index, inplace = True)
```

```python
df
```

```python
df['Turnover'] = df['Avg. price']*df['Qty.']
```

```python
df
```

```python
df['Brokerage'] = df['Turnover']*0.0003
```

```python
df['Brokerage'] = np.where(df['Brokerage']>= 20,20,df['Brokerage'])
```

```python
df
```

```python
df['STT/CTT'] = np.where(df['Type'] == "SELL",df['Turnover']*0.00025,0).round(2)
```

```python
df
```

```python
df['ETC'] = (df['Turnover']*0.0000325).round(2)
```

```python
df
```

```python
df['SEBI'] = (df['Turnover']*(10/10000000)).round(2)
```

```python
df
```

```python
df['GST'] = ((df['Brokerage'] + df['ETC'] + df['SEBI'])*(0.18)).round(2)
```

```python
df
```

```python
df['Stamp charges'] =  np.where(df['Type'] == "BUY",df['Turnover']*0.00003,0).round(2)
```

```python
df
```

```python
df['Stamp charges'] = np.where(df['Stamp charges']>= 300,300,df['Stamp charges'])
```

```python
df
```

```python
df['Total charges'] = df['Brokerage'] + df['STT/CTT'] + df['ETC'] + df['SEBI'] + df['GST'] + df['Stamp charges']
```

```python
df
```

```python
All_Transaction = df
```

```python
All_Transaction
```

```python
writer = pd.ExcelWriter('Kite PnL Solution.xlsx')
```

```python
All_Transaction.to_excel(writer,sheet_name="Individual Transaction Charges")
```

```python
writer.save()
```

```python
stokwise_typewise = pd.DataFrame(df.groupby(['Instrument','Type'],as_index = False).sum())
```

```python
stokwise_typewise
```

```python
st_typewise = pd.DataFrame(df.groupby(['Instrument','Type'],as_index = False).agg({"Qty.":['sum'],'Avg. price':['mean'],'Turnover':['sum'],'Brokerage':['sum'],'STT/CTT':['sum'],'ETC':['sum'],'SEBI':['sum'],'GST':['sum'],'Stamp charges':['sum'],'Total charges':['sum']}))
```

```python
st_typewise
```

```python
stokwise_typewise.to_excel(writer,sheet_name="Stockwise & Type wise")
```

```python
st_typewise.to_excel(writer,sheet_name="Stockwise & Type wise summary")
```

```python
writer.save()
```

```python
DF = pd.DataFrame(df['Instrument'].unique(),columns = ['Instrumnet'])
```

```python
DF
```

```python
DF['Gross PnL'] = pd.DataFrame(stokwise_typewise.groupby('Instrument')['Turnover'].diff().dropna().reset_index(drop = True))
```

```python
DF
```

```python
DF['Total Charge'] = pd.DataFrame(stokwise_typewise.groupby('Instrument')['Total charges'].sum().dropna().reset_index(drop = True))
```

```python
DF
```

```python
DF['Net PnL'] = DF['Gross PnL'] - DF['Total Charge']
```

```python
DF
```

```python
DF['% charge'] = (DF['Total Charge']/DF['Gross PnL']*100).round(2)
```

```python
DF
```

```python
Total_Gross = DF['Gross PnL'].sum()
```

```python
Total_Gross
```

```python
Total_charges = DF['Total Charge'].sum()
```

```python
Total_charges
```

```python
Net_PnL = DF["Net PnL"].sum().round(2)
```

```python
Net_PnL
```

```python
charge = (Total_charges/Total_Gross*100).round(2)
```

```python
charge
```

```python
Total = {'Instrumnet':'Total','Gross PnL':Total_Gross,'Total Charge':Total_charges,'Net PnL':Net_PnL,'% charge':charge}
```

```python
Total_summary = DF.append(Total,ignore_index = True)
```

```python
Total_summary
```

```python
Total_summary.to_excel(writer,sheet_name="Stockwise Summary")
```
